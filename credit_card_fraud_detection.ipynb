```python
#DATA TRANSFORMATION - Normalization and standardization
#importing from scikit learn
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
#Normalizing numeric features(time and amount)
scaler = MinMaxScaler()

data[['Time','Amount']] = scaler.fit_transform(data[['Time','Amount']])
```

```python
#Standardization of features time and amount
standard_scaler = StandardScaler()

data[['Time','Amount']] = standard_scaler.fit_transform(data[['Time','Amount']])

#display data after standardization
data
```

```python
#Scaling PCA-transformed features(V1,V2,V3,....)
pca_features = []
for i in range(1,29):
    pca_features.append('V'+ str(i))
    i=i+1

pca_features
```

```python
#standardizing PCA transformed features
pca_scaler = StandardScaler()
data[pca_features] = pca_scaler.fit_transform(data[pca_features])
#disply scaled dataset
data
```

```python
#converting time to hours of day
data['Hour']= (data['Time']/3600)
data['Hour'].describe()
```

```python
#Log transformation of Amount
data['log_Amount'] = np.log1p(data['Amount'])
data['log_Amount'].describe()
```

```python
data[['Hour','log_Amount']] = StandardScaler().fit_transform(data[['Hour','log_Amount']])
```

```python
for i in range(1,29,2):
    data[f'InteractionV{i}_V{i+1}']=data[f'V{i}']*data[f'V{i+1}']

data['InteractionV1_V2']
```

```python
#Model development
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,auc
```

```python
#splitting data into features and target variable
x = data.drop('Class',axis=1)
y = data['Class']
```

```python
#splitting data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Using logistic Regreesion model for evaluation

#importing necessary libraries
from sklearn.linear_model import LogisticRegression

#Training the model
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)

#prediciting the results
logpred = logmodel.predict(x_test)

#Evaluation
print(classification_report(y_test,logpred))
print(confusion_matrix(y_test,logpred))
#plotting ROC curve
logpred_prob = logmodel.predict_proba(x_test)[:, 1]

fpr, tpr , thresholds = roc_curve(y_test, logpred_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#Using Random Forest model for evaluation

#importing necessary libraries
from sklearn.ensemble import RandomForestClassifier

#Training the model
ranmodel = RandomForestClassifier(n_estimators=100,random_state=42)
ranmodel.fit(x_train,y_train)

#prediciting the results
ranpred = ranmodel.predict(x_test)

#Evaluation
print(classification_report(y_test,ranpred))
print(confusion_matrix(y_test,ranpred))

#plotting ROC curve
ranpred_prob = ranmodel.predict_proba(x_test)[:, 1]

fpr, tpr , thresholds = roc_curve(y_test, ranpred_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#Using Gradient Boosting model for evaluation

#importing necessary libraries
from sklearn.ensemble import GradientBoostingClassifier

#Training the model
gdmodel = GradientBoostingClassifier(n_estimators=100,random_state=42)
gdmodel.fit(x_train,y_train)

#prediciting the results
gdpred = gdmodel.predict(x_test)

#Evaluation
print(classification_report(y_test,gdpred))
print(confusion_matrix(y_test,gdpred))

#plotting ROC curve
gdpred_prob = gdmodel.predict_proba(x_test)[:, 1]

fpr, tpr , thresholds = roc_curve(y_test, gdpred_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#Deciding the Best Model
#Precision: High precision is preferred since we want minimal false positives.RandomForest model has the highest.
#Recall: High recall is preferred since we want to minimize false negatives.RandomForest has the highest.
#F1-Score: This provides a balance between precision and recall.RandomForest model has the highest.
#ROC-AUC Score: This gives an overall measure of model performance. LogisticRegression has the highest.
```

```python
#Addressing class imbalance using SMOTE
#importing necessary libraries
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
x_smote,y_smote = sm.fit_resample(x_train,y_train)

#Scaling features
scaler = StandardScaler()
x_smote = scaler.fit_transform(x_smote)
x_test = scaler.transform(x_test)
```

```python
#Comparing before and after using SMOTE
fig,ax = plt.subplots(2,1,sharex=True)
ax[0].hist(y_train)
ax[0].set_ylabel('count')
ax[1].hist(y_smote)
ax[1].set_xlabel('class')
ax[1].set_ylabel('count')
plt.title('Before and after balancing')
plt.show()
```

```python
#Logistic Regression with SMOTE
#Using logistic Regreesion model for evaluation

#importing necessary libraries
from sklearn.linear_model import LogisticRegression

#Training the model
logmodel_sm = LogisticRegression()
logmodel_sm.fit(x_smote,y_smote)

#prediciting the results
logpred_sm = logmodel_sm.predict(x_test)

#Evaluation
print(classification_report(y_test,logpred_sm))
print(confusion_matrix(y_test,logpred_sm))
#plotting ROC curve
logpred_prob_sm = logmodel_sm.predict_proba(x_test)[:, 1]

fpr, tpr , thresholds = roc_curve(y_test, logpred_prob_sm)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
#Random Forest with SMOTE
#Using Random Forest model for evaluation

#importing necessary libraries
from sklearn.ensemble import RandomForestClassifier

#Training the model
ranmodel_sm = RandomForestClassifier(n_estimators=100,random_state=42)
ranmodel_sm.fit(x_train,y_train)

#prediciting the results
ranpred_sm = ranmodel_sm.predict(x_test)

#Evaluation
print(classification_report(y_test,ranpred_sm))
print(confusion_matrix(y_test,ranpred_sm))

#plotting ROC curve
ranpred_prob_sm = ranmodel_sm.predict_proba(x_test)[:, 1]

fpr, tpr , thresholds = roc_curve(y_test, ranpred_prob_sm)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
